```python
def plott(images):
    # Display the first ten images
    fig, axes = plt.subplots(1, 10, figsize=(10, 10))  # Create a 2x5 grid for displaying 10 images

    for i, ax in enumerate(axes.ravel()):
        ax.imshow(images[i], cmap='gray')  # Display each image in grayscale
        # ax.set_title(f'Image {i+1}')
        ax.axis('off')  # Turn off axis labels

    plt.tight_layout()
    plt.show()
# Reshape the 'd' array into (30000, 24, 24)
new_shape = (30000, 24, 24)
d_reshaped = d.reshape(new_shape)

# Extract the first ten (24, 24) arrays
images = d_reshaped[0:10]
plott(images)

# np.tanh

# Extract the first ten (24, 24) arrays
t_images = np.tanh(images)
plott(t_images)
```
